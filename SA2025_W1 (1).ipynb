### 1. What is name of car that has the highest horsepower?  

```python
import pandas as pd

df = pd.read_csv('Cars.csv')  

print(df.columns)  # Check all columns loaded
# Find the row with the highest horsepower
max_hp_row = df.loc[df['horsepower'].idxmax()]

# Extract the car name from that row
car_name = max_hp_row['name']

print("Car with highest horsepower:", car_name)


```

### 2. How many cars have mpg  35?

```python
count = df[df['mpg'] >= 35].shape[0]
print("Number of cars with mpg >= 35:", count)

```

### 3. What is the most common origin for cars with horsepower > 100 and weight < 3000?

```python
filtered_cars = df[(df['horsepower'] > 100) & (df['weight'] < 3000)]

if not filtered_cars.empty:
    most_common_origin = filtered_cars['origin'].mode()[0]
    print("Most common origin for cars with hp > 100 and weight < 3000:", most_common_origin)
else:
    print("No cars found with horsepower > 100 and weight < 300.")


```

### 4. What is the mean acceleration of cars from Japan? (rounded to 2 decimals)

```python
mean_accel = df[df['origin'] == 'japan']['acceleration'].mean()
mean_accel_rounded = round(mean_accel, 2)
print("Mean acceleration of cars from Japan:", mean_accel_rounded)

```

### 5. Which year had the highest average mpg?

```python
# Group by model_year and calculate average mpg
avg_mpg_by_year = df.groupby('model_year')['mpg'].mean()

# Find the year with the highest average mpg
year_with_highest_avg_mpg = avg_mpg_by_year.idxmax()

print("Year with the highest average mpg:", year_with_highest_avg_mpg)

```

### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
# Calculate median mpg
median_mpg = df['mpg'].median()
# Filter cars with mpg above median
filtered_cars = df[df['mpg'] > median_mpg].copy()
# Calculate horsepower-to-weight ratio
filtered_cars['hp_per_weight'] = filtered_cars['horsepower'] / filtered_cars['weight']
# Find the index of the car with the max hp_per_weight ratio
best_ratio_index = filtered_cars['hp_per_weight'].idxmax()
# Get that row
best_car = filtered_cars.loc[best_ratio_index]
print("Car with best horsepower-to-weight ratio (mpg above median):")
print(best_car)
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Calculate average mpg per year and origin
avg_mpg = df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()
# Set plot size and style
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")
# Plot multi-line plot with seaborn lineplot
sns.lineplot(data=avg_mpg, x='model_year', y='mpg', hue='origin', marker='o')
# Add title and labels
plt.title('Evolution of Average MPG Over Years by Region')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.legend(title='Origin')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
plt.figure(figsize=(12, 8))
sns.set_style("whitegrid")
# Create scatterplot
scatter = sns.scatterplot(
    data=df,
    x='horsepower',
    y='weight',
    hue='origin',
    size='mpg',
    sizes=(20, 200),  # adjust point size range
    hue_order=['japan', 'europe', 'usa'],
    palette='Set2',
    alpha=0.7
)
# Add titles and labels
plt.title('Scatterplot of Horsepower vs Weight by Origin and MPG Size', fontsize=16)
plt.xlabel('Horsepower', fontsize=14)
plt.ylabel('Weight', fontsize=14)
# Improve legend
plt.legend(title='Origin', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
